```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sinh(np.sin(x / 5)**2)

x_nodes = np.arange(0, 10, 0.05)
x_1_nodes = np.arange(0.025, 10, 0.05)

print(x_nodes, x_1_nodes)

y_nodes = f(x_nodes)
y_1_nodes = []

def lagrange_interpolation(x_values, y_values, x):
    n = len(x_values)
    result = 0
    for i in range(n):
        term = y_values[i]
        for j in range(n):
            if i != j:
                term *= (x - x_values[j]) / (x_values[i] - x_values[j])
        result += term
    return result

for x in x_1_nodes:
    y_1_nodes.append(lagrange_interpolation(x_nodes, y_nodes, x))


x_1_nodes = x_1_nodes[50:-50]
y_1_nodes = y_1_nodes[50:-50]


plt.figure(figsize=(10, 6))
plt.plot(x_nodes, y_nodes, 'b-', label='Исходные данные')
plt.plot(x_1_nodes, y_1_nodes, 'r--', label='Интерполяция')


plt.xlabel('x')
plt.ylabel('y')
plt.title('График интерполяции Лагранжа')
plt.legend()
plt.grid(True)
plt.show()

true_values = f(x_1_nodes)

errors = np.abs(true_values - y_1_nodes)
max_error = np.max(errors)

print(max_error)
```

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return np.sinh(np.sin(x / 5)**2)


def chebyshev_nodes(n, a, b):
    k = np.arange(1, n + 1)
    nodes = np.cos((2 * k - 1) * np.pi / (2 * n))
    nodes = 0.5 * (a + b + (b - a) * nodes)
    return nodes


n = 100
a, b = 0, 10


chebyshev_x_1 = chebyshev_nodes(n, a, b)
chebyshev_y_1 = f(chebyshev_x_1)


chebyshev_x_2 = []
chebyshev_y_2 = []


for i in range(0, len(chebyshev_x_1) - 1):
    chebyshev_x_2.append((chebyshev_x_1[i] + chebyshev_x_1[i+1]) / 2)
chebyshev_x_2.append(chebyshev_x_1[-1])  # Добавление последней точки


def lagrange_interpolation(x_values, y_values, x):
    n = len(x_values)
    result = 0
    for i in range(n):
        term = y_values[i]
        for j in range(n):
            if i != j:
                term *= (x - x_values[j]) / (x_values[i] - x_values[j])
        result += term
    return result


for x in chebyshev_x_2:
    chebyshev_y_2.append(lagrange_interpolation(chebyshev_x_1, chebyshev_y_1, x))


chebyshev_x_2 = np.array(chebyshev_x_2)


plt.figure(figsize=(10, 6))
plt.plot(chebyshev_x_1, chebyshev_y_1, 'b-', label='Исходные данные')
plt.plot(chebyshev_x_2, chebyshev_y_2, 'r--', label='Интерполяция')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График интерполяции с точками Чебышева')
plt.legend()
plt.grid(True)
#plt.scatter(chebyshev_x_2, chebyshev_y_2, color='red')
#plt.scatter(chebyshev_x_1, chebyshev_y_1,  color='blue')
plt.show()


print("Исходные узлы Чебышева (chebyshev_x_1): \n", chebyshev_x_1)
print("Половинные узлы Чебышева (chebyshev_x_2): \n", chebyshev_x_2)

true_values2 = f(chebyshev_x_2)
errors = np.abs(true_values2 - chebyshev_y_2)
max_error = np.max(errors)

print("Максимальная ошибка интерполяции:", max_error)
```
